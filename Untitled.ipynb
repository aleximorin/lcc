```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat

%matplotlib inline
```

```python
path = 'data/Findelen_2207_datacube_700_2.mat'
cube = loadmat(path, mat_dtype=True, squeeze_me=True, struct_as_record=False)
data = cube['datacube_2'].datacube
```

```python
im = data[:, :, 100]
vmax = np.nanquantile(np.abs(im), 0.99)

plt.imshow(im, aspect='auto', cmap='RdBu', vmin=-vmax, vmax=vmax)
```

```python
im2 = im[:, 85:]
vmax = np.nanquantile(np.abs(im2), 0.99)

plt.imshow(im2, aspect='auto', cmap='RdBu', vmin=-vmax, vmax=vmax)
```
